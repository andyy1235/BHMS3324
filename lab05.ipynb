### Line Styles

You can also mark the data points by specifying the marker type in the `marker` argument of the `plot()` function.
The table below shows the most commonly used markers.

| Marker | Description | | Marker | Description|
|:-:|:-:|:-:|:-:|:-:|
|'o' | Circle| | '\*' | Start|
|'.' | Point| | ',' | Pixel|
|'x' | X | | '+' | Plus|
|'D' | Diamond| | 'P' | Pentagon|
|'^' | Triangle Up| | 'v' | Triangle Down|


```python
x = [0, 1]
y = [3, 10]

plt.title("SML")
plt.xlabel("Beta")
plt.ylabel("Return (%)")

plt.plot(x, y, marker='o' )
```

You can also specify the line style by "Format String". "Format String" is a shortcut string notation parameter, and the parameter is also called `fmt`. The syntax of `fmt` is *`marker | line | color`*


| Line | Description | | Color | Description|
|:-:|:-:|:-:|:-:|:-:|
|'-' | Solide line| | 'r' | Red|
|':' | Dotted line| | 'g' | Green|
|'--' | Dashed line | | 'b' | Blue|
|'-.' | Dashed-dotted line| | 'c' | Cyan|
|- | -| | 'm' | Magenta|
|- | -| | 'y' | Yellow|
|- | -| | 'k' | Black|

For example, if we want to plot a red dotted line with marker of 'star', we set the `fmt` as:

`'*:r'`

### Using NumPy in Data Science

Programmers working with data science projects usually work with the `Numpy` library.

We use `Numpy` for working with arrays. `Numpy` facilitates us to work in the areas of linear algebra and matrices.

The following example uses the NumPy function to plot a linear regression line.


```python
import numpy as np

# declare the data as Numpy array
price = np.array([5,6,9,8.5,7,11,5.5,23])
size = np.array([400,430,600,650,500,800,500,1000])

plt.scatter(size, price)

# use numpy's polyfit function to return the slope and intercept fit
m, c = np.polyfit(size, price, 1)

# x axis is the point in the 'size' array, we obtain the respective values of y by using the linear equation of y=mx+c.
plt.plot(size, m*size + c)

print("The slope of the line is {slope:.2f}, and the y-intercept is {intercept:.2f}.".format(slope=m, intercept=c))
```

### Obtaining Stock Data

Now, let's obtain the 'Apple Inc' Stock data from 1 Jan 2020 to 31 Dec 2020.


```python
import yfinance as yf

ticker = "AAPL"
start = "2020-01-01"
end = "2020-12-31"

data_df = yf.download(ticker, start, end)

data_df.head()
```

#### Now, let's calcuate the daily return of the stock.

We can use the `pct_change()` function provided by pandas to calculate the daily, monthly or annual return.

```python
print(data_df.columns)
data_df["Close"].pct_change(1)  # 1 is for daily, 21 is for monthly, 252 is for annually.
```

```python
return_df = pd.DataFrame()
return_df['Price'] = data_df["Close"]
return_df['Return'] = return_df['Price'].pct_change(1)
return_df.head()
```

```python
# Remove the all the NA row

return_df = return_df.dropna()
return_df.head()
```

```python
return_df['Price'].plot()
```
